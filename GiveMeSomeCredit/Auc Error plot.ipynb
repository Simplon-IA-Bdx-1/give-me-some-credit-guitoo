```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
deepnet=read_csv('./Deepnet.csv',index_col=0)
ensemble=read_csv('./Ensemble.csv',index_col=0)
```

```python
deepnet
```

```python
ensemble
```

```python
ensemble_mean = ensemble.mean(skipna = True) 
ensemble_std = ensemble.std(skipna = True)
deepnet_mean = deepnet.mean(skipna = True) 
deepnet_std = deepnet.std(skipna = True)
percent=[x for x in range(10,101,10)]
```

```python
deepnet_mean-deepnet_std
```

```python
#xfit = np.linspace(10, 100, 100)
plt.plot(percent, ensemble_mean, color='blue')
plt.fill_between(percent, ensemble_mean - ensemble_std, ensemble_mean + ensemble_std, color='blue', alpha=0.2)
plt.plot(percent, deepnet_mean, color='red')
plt.fill_between(percent, deepnet_mean - deepnet_std, deepnet_mean + deepnet_std, color='red', alpha=0.2)
plt.show()
```
