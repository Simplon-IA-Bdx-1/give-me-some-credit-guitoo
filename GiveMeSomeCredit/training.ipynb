```python
from pandas import DataFrame, read_csv, to_numeric, Series
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt
from bigml.api import BigML
import kaggle
from pprint import pprint

%matplotlib inline
```

```python
compression = True  #Do we compress the csv files?

version='v1.3'

big_ml_project = 'project/5d94a428eba31d460c00023f'
```

```python
if compression:
    compress='.bz2'
else:
    compress=''
  
filename_fulltrain = 'gmsc-fulltrain-' + version + '.csv'+ compress
filename_train = 'gmsc-train80-' + version + '.csv'+ compress
filename_valid = 'gmsc-valid20-' + version + '.csv'+ compress
filename_test = 'gmsc-test-' + version + '.csv'+ compress

filename_bigml_ids = 'bigml-' + version + '.csv'+ compress
```

```python
#bigml_ids = read_csv(filename_bigml_ids, squeeze=True, index_col=0)    
```

## Send the data to BigML and create an ensemble model

```python
api = BigML(project=big_ml_project)

train_src = api.create_source(filename_train)
api.ok(train_src)
valid_src = api.create_source(filename_valid)
api.ok(valid_src)
test_src = api.create_source(filename_test)
api.ok(test_src)

print("Sources created")
```

```python
train_ds = api.create_dataset(train_src)
api.ok(train_ds)
valid_ds = api.create_dataset(valid_src)
api.ok(valid_src)
test_ds = api.create_dataset(test_src)
api.ok(test_src)
print("Data sets created")
```

```python
model_args= {"objective_field": "SeriousDlqin2yrs"}
model = api.create_ensemble(train_ds, model_args)
#model = api.create_deepnet(train_ds, model_args)
api.ok(model)
print("Model created")
```

Save BigML ids in a csv file

```python
bigml_ids = {}
bigml_ids['training source']=train_src['resource']
bigml_ids['validation source']=valid_src['resource']
bigml_ids['test source']=test_src['resource']

bigml_ids['training dataset']=train_ds['resource']
bigml_ids['validation dataset']=valid_ds['resource']
bigml_ids['test dataset']=test_ds['resource']

bigml_ids['model']=model['resource']

pprint(bigml_ids)

Series(bigml_ids).to_csv(filename_bigml_ids, header=False)
```
